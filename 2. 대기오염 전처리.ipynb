```python
data_columns = df.iloc[:, 6:].columns
for column in data_columns:
    median_value = df[column].median()
    df.loc[df[column] == -999, column] = median_value
```

```python
df['측정일시'] = pd.to_datetime(df['측정일시'])
```

```python
df['측정연월'] = pd.to_datetime(df['측정일시'].dt.year.astype(str) + '-' + df['측정일시'].dt.month.astype(str))
```

```python
df = df.drop('측정일시', axis=1)
```

```python
data = df.groupby(['측정연월', '측정소코드', '측정소명', '시도', '도시', '군구'], as_index=False).mean()
```

```python
data
```

```python
data = data[(data['측정연월'].dt.year < 2023)]    #2023년도 제외
```
