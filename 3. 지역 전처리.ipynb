```python
import os
import pandas as pd
import numpy as np
```

```python
disease_df = pd.read_excel('data/질병데이터_전처리.xlsx')
air_df = pd.read_excel('data/대기오염_전처리.xlsx')
```

```python
disease_df.head()
```

```python
disease_df.시도명.unique()
```

```python
disease_df[disease_df['시도명'] == '대구광역시']
```

```python
sido_list = disease_df.시도명.unique()
```

```python
def find_sido(x):
    for sido in sido_list:
        if sido.startswith(x):
            return sido
    return x
```

```python
air_df.head()
```

```python
air_df.시도.unique()
```

```python
air_df['시도명'] = air_df['시도'].apply(find_sido)
```

```python
disease_df.시도명.unique()
```

```python
air_df.시도명.unique()
```

```python
air_df.loc[air_df['시도명'] == '경남', '시도명'] = '경상남도'
air_df.loc[air_df['시도명'] == '전북', '시도명'] = '전라북도'
air_df.loc[air_df['시도명'] == '전남', '시도명'] = '전라남도'
air_df.loc[air_df['시도명'] == '경북', '시도명'] = '경상북도'
air_df.loc[air_df['시도명'] == '충북', '시도명'] = '충청북도'
air_df.loc[air_df['시도명'] == '충남', '시도명'] = '충청남도'
```

```python
air_df.시도명.unique()
```

```python
disease_df.시군구명.unique()
```

```python
disease_df[disease_df['시도명'] == '세종특별자치시'].시군구명.unique()
```

```python
disease_df.loc[disease_df['시도명'] == '세종특별자치시', '시군구명'] = '세종시'
```

```python
disease_df['new_시군구명'] = disease_df['시군구명'].apply(lambda x: x.split()[0])
```

```python
disease_df['new_시군구명'].unique()
```

```python
air_df.군구.unique()
```

```python
def find_address(x):
    for address in disease_df['new_시군구명'].unique():
        if (x == address) | (address.startswith(x)):
            return address
    return ''
```

```python
air_df['시군구명'] = air_df['군구'].apply(find_address)
```

```python
len(air_df)
```

```python
len(air_df[air_df['시군구명'] == ''])
```

```python
# 군위군만 예외처리
air_df[air_df['시군구명'] == '군위군']  # 경상북도에서 대구광역시로 변경
air_df.loc[air_df['시군구명'] == '군위군', '시도명'] = '대구광역시'
```

```python
disease_df
```

```python
disease_df['new_지역명'] = disease_df['시도명'] + ' ' + disease_df['new_시군구명']
```

```python
air_df.head()
```

```python
air_df['new_지역명'] = air_df['시도명'] + ' ' + air_df['시군구명']
```

```python
new_disease_df = disease_df.groupby(['요양개시연월', 'new_지역명'], as_index=False)[['아토피', '비염', '천식']].sum()
```

```python
new_disease_df.head()
```

```python
air_df.columns
```

```python
new_air_df = air_df.groupby(['측정연월', 'new_지역명'], as_index=False)[['아황산가스', '미세먼지', '오존', '이산화질소', '일산화탄소', '초미세먼지']].mean()
```

```python
new_air_df.head()
```

```python
new_disease_df = new_disease_df.rename(columns={'요양개시연월': '측정연월'})
```

```python
len(new_air_df)
```

```python
len(new_disease_df)
```

```python
new_df = pd.merge(new_disease_df, new_air_df, on=['측정연월', 'new_지역명'], how='inner')
```
